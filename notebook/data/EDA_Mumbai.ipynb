```python
## Feature Engineering

import seaborn as sns
location_count = df2['Location'].value_counts()
plt.figure(figsize=(12, 70))
sns.barplot(x=location_count.values, y=location_count.index, orient='h')
plt.show()
```

```python
df2.Location = df2.Location.apply(lambda x: x.strip())
location_count.sort_values(ascending=False)
```

```python
len(location_count[location_count<=10])
```

```python
location_count_lessthan10 = location_count[location_count<=10]
sns.histplot(location_count_lessthan10)
```

```python
df3 = df2.copy()    ##Replaced datapoints having locations whose count is less than 10 with other
df3.Location = df3.Location.apply(lambda x: 'other' if x in location_count_lessthan10 else x)
len(df3.Location.unique())
```

```python
df3.head(10)
```

```python
area_per_bhk = df3['Area']/df3['size']
area_per_bhk_df = pd.DataFrame(area_per_bhk)
area_per_bhk_df.describe()
```

```python
df4 = df3.copy()                          ## df4 is defined for adding area/BHK
df4['area_per_bhk'] = area_per_bhk
```

```python
df5 = df4.copy()                          ## df5 is defined for adding price/area
price_per_area = df4['Price']/df4['Area']
price_per_area_df = pd.DataFrame(price_per_area)
price_per_area_df.describe()
```

```python
df5 ['price_per_area'] = price_per_area
df5.head()
```

```python
##Outlier Removal

df6 = df5.copy()
print(df5[df5['area_per_bhk']<350])
df6 = df5[~(df5['area_per_bhk']<350)]    ## removed rows having area/BHK less than 350
df6.shape
```

```python
def remove_ppa_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        M = np.mean(subdf.price_per_area)
        SD = np.std(subdf.price_per_area)
        reduced_df = subdf[(subdf.price_per_area>(M-SD)) & (subdf.price_per_area<=(M+SD))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df7 = remove_ppa_outliers(df6)       ## df7 is defined for removing outliers in distribution of price/area for a specific location(within 1 S.D.)
df7.shape
```

```python
def areavspriceplot_location(df, location):
    BHK1 = df[(df['Location'] == location) & (df['size'] == 1)]
    BHK2 = df[(df['Location'] == location) & (df['size'] ==2)]
    BHK3 = df[(df['Location'] == location) & (df['size'] ==3)]
    BHK4 = df[(df['Location'] == location) & (df['size'] ==4)]
    plt.rcParams['figure.figsize'] = (15,10)
    plt.scatter(BHK1.Area, BHK1.Price, color='blue', label='1 BHK', s=30)
    plt.scatter(BHK2.Area, BHK2.Price, marker='*', color='yellow', label='2 BHK', s=30)
    plt.scatter(BHK3.Area, BHK3.Price, marker='+', color='red', label='3 BHK', s=30)
    plt.scatter(BHK4.Area, BHK4.Price, marker='o', color='green', label='4 BHK', s=30)
    plt.xlabel("Total Area (in sq.ft.)")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()
    return plt.show()

areavspriceplot_location(df7, "Kharghar")
```

```python
def remove_bhkprice_outliers(df):       #assumed that bigger size bhk will cost more than smaller ones
    remove_indices = np.array([])       #for the same location and approximately same area.
    for l, location_df in df.groupby('Location'):
        bhk_price_stats = {}
        for b, bhk_df in location_df.groupby('size'):
            bhk_price_stats [b]={
                'mean': np.mean(bhk_df.price_per_area),
                'SD' : np.std(bhk_df.price_per_area),
                'count' : bhk_df.shape[0]
            }
        for r, bhk_df in location_df.groupby('size'):
            stat = bhk_price_stats.get(r-1)
            if stat and stat['count']>6:
                remove_indices = np.append(remove_indices, bhk_df[bhk_df.price_per_area<(stat['mean'])].index.values)
    print(bhk_price_stats)
    return df.drop(remove_indices, axis='index')

df8 = remove_bhkprice_outliers(df7)         #df8 is defined for removing BHK outliers for a specific location
```
